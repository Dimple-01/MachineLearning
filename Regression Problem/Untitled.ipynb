Gradient Descent

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x= np.arange(10) # generate numbers from 0 to 10
y= (x-5)**2
print(x,y)
```

Goal
given a f(x) we want to find value of x that minimizes f.

Visualization

```python
#plt.style.use("seaborn")
plt.plot(x,y)
plt.xlabel("x")
plt.ylabel("F(x)")
plt.show()
```
